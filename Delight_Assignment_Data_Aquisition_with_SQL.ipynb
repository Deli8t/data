## **Loading our Data Set**

```python
# Loading the pandas library
# ---
#
import pandas as pd

# Step 1: Reading the dataset from the given URL using the pandas library.
# and storing the dataset in a variable called 'cities'.
# ---
#
accidents_df = pd.read_csv('/content/traffic_accidents.csv') 

# Step 2: We check whether there's a table with the intended new table name.
# ---
# 
%sql DROP TABLE if EXISTS accidents_df;

# Step 3: We create a new table called 'cities' within our in memory database. The persist 
# command takes the table from the 'cities' variable and stores it in the database.
# ---
#
%sql PERSIST accidents_df;

# Step 4: We  select the first five records of the created cities table.
# ---
#
%sql SELECT * FROM accidents_df LIMIT 5;
```

```python
#Display a list of 5 makes that had the highest number of crashes

%%sql
SELECT MAKE, SUM(MODEL) FROM accidents_df
GROUP BY MAKE
ORDER BY SUM(MODEL) DESC
LIMIT 5
```

```python
#Most personal vehicles involved in crashes were traveling towards which direction?


%%sql
SELECT VEHICLE_USE, TRAVEL_DIRECTION, COUNT(TRAVEL_DIRECTION) FROM accidents_df
WHERE VEHICLE_USE = 'PERSONAL'
GROUP BY TRAVEL_DIRECTION
ORDER BY COUNT(TRAVEL_DIRECTION)
```

```python
#How many passengers were involved in personal cars were in accidents?

%%sql
SELECT VEHICLE_USE, SUM(NUM_PASSENGERS) FROM accidents_df
WHERE VEHICLE_USE = 'PERSONAL'

```

```python
#How many passengers were involved in accidents while traveling in personal Sport Utility vehicles?

%%sql
SELECT VEHICLE_TYPE, SUM(NUM_PASSENGERS) FROM accidents_df
WHERE  VEHICLE_TYPE='SPORT UTILITY VEHICLE (SUV)'

```

```python
#Compute the total number of Mercedez-Benz vehicles involved in accidents while being parked

%%sql
SELECT MAKE, MANEUVER, SUM(MANEUVER) FROM accidents_df
WHERE MAKE = 'MERCEDES-BENZ' AND MANEUVER ='PARKED'
```

```python
#Compute the total occupant count per vehicle count.

%%sql
SELECT MAKE, SUM(OCCUPANT_CNT) FROM accidents_df
GROUP BY MAKE
ORDER BY SUM(OCCUPANT_CNT) DESC
```

```python
#Sort the number of passengers in descending order by travel direction taking into account travel direction. 
#Return travel direction and no. of passengers.

%%sql
SELECT TRAVEL_DIRECTION, SUM(NUM_PASSENGERS) FROM accidents_df
GROUP BY TRAVEL_DIRECTION
ORDER BY SUM(NUM_PASSENGERS) DESC
```

```python
#Select the top 5 vehicle models involved in the accident with no defect, order them by make, and show their maneuver.

%%sql
SELECT MODEL, COUNT(MODEL), MAKE ,VEHICLE_DEFECT,MANEUVER FROM accidents_df
WHERE VEHICLE_DEFECT = 'NONE'
GROUP BY MODEL
ORDER BY COUNT(MAKE) DESC
LIMIT 5
```

```python
#Which ridesharing vehicles in an accident had the most number of passengers?
#Return model, and the number of passengers.

%%sql
SELECT   MODEL, COUNT(NUM_PASSENGERS) FROM accidents_df
WHERE VEHICLE_USE= 'RIDESHARE SERVICE'
GROUP BY MODEL
ORDER BY COUNT(NUM_PASSENGERS) DESC

```
